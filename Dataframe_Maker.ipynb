```python
import pandas as pd
import glob
```

```python
pattern = 'data/*.csv'
csv_files = glob.glob(pattern)
```

```python
csv_files
```

```python
pattern1 = 'data/Course ID/*.csv'
cols2add = glob.glob(pattern1)
```

```python
pd.read_csv('data/df_0k_10k.csv', index_col=0)
```

```python
frames = []
for csv in csv_files:
    df = pd.read_csv(csv, index_col=0)
    frames.append(df)

reviews = pd.concat(frames).drop(['course_id', 'course_name'], axis=1).reset_index()
```

```python
frames = []
for cols in cols2add:
    df = pd.read_csv(cols, index_col=0)
    frames.append(df)
    
course_id = pd.concat(frames).reset_index()
```

```python
df = pd.concat([reviews, course_id], axis=1).drop('index', axis = 1)
```

```python
df.to_csv('dataframe.csv')
```
